```python
import glob, os, re
```

```python
namePattern = re.compile(r'^([^\d]+)(\d\d\d)([^\d]*?)(\d+)$')

def fixName(oldName):
    m = re.match(namePattern, oldName)
    (a, b, c, d) = (m.group(1), m.group(2), m.group(3), m.group(4))
    b = '{:03d}'.format(int(b))
    if (c == ''):
        c = '0'
    d = '{:03d}'.format(int(d))
    return ''.join((a, b, c, d)).upper();

def correct_column_file(filename):
    name = os.path.splitext(os.path.basename(filename))[0]
    corrected_name = fixName(name)
    if name == corrected_name:
        print '{name} needs no correction'.format(**locals())
    else:
        dest = os.path.dirname(filename) + '/' + corrected_name + '.numpy'
        if os.path.exists(dest):
            print '{name} has already been corrected'.format(**locals())
        else:
            os.symlink(os.path.basename(filename), dest)
            print 'Symlinked {corrected_name} to {name}'.format(**locals())


# correct_column_file('columncache/census2000_block2010/p012d006.numpy')
```

```python
for filename in glob.glob('columncache/*_block2010/*.numpy'):
    correct_column_file(filename)
```
