```python
countries_sheet_export_url = 'https://docs.google.com/spreadsheets/u/0/d/11Q9Gq-BNTElOsA4u5BdIcM8CWG-eWRT8c3I_8AQ03CM/export?format=csv&id=11Q9Gq-BNTElOsA4u5BdIcM8CWG-eWRT8c3I_8AQ03CM&gid=0'
countries_sheet_data = [row for row in csv.reader(urllib2.urlopen(countries_sheet_export_url))]
country_aliases = {}
for place in countries_sheet_data[1:]:
    alpha3 = place[0]
    for alias in place:
        country_aliases[alias.lower()] = alpha3
```

```python
def map_country_to_alpha3(name):
    try:
        return country_aliases[name.lower()]
    except:
        return None
    
map_country_to_alpha3('af')
```

```python
def add_country_mapping(alpha3, name):
    print 'Adding mapping "%s"->"%s"' % (name, alpha3)

def add_gapminder_country(gc):
    name = gc['properties']['gapminder_name']
    alpha2 = gc['properties']['iso_3136']
    if not alpha2:
        print 'Gapminder gives no alpha2 for "%s", skipping' % name
        return
    alpha3 = map_to_alpha3(alpha2)
    assert alpha3
    if not map_country_to_alpha3(name):
        add_country_mapping(alpha3, name)
    elif map_country_to_alpha3(name) == alpha3:
        print 'Gapminder mapping already exists "%s"->"%s"' % (name, map_country_to_alpha3(name))
    else:
        raise Exception('Oops, country mapping claims "%s"->"%s" but gapminder claims "%s"->"%s"->"%s"' %
                        (name, map_country_to_alpha3(name), name, alpha2, alpha3))
                        

for gapminder_country in gj['features']:
    add_gapminder_country(gapminder_country)


```
